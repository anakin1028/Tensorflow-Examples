```python
# Compose a deep bidrectional rnn model
# The first two layers are Convolutional Nueral Networks
# Then follows the Bi-Directional Recurrent Neural Networks

import tensorflow as tf
import numpy as np
import matplotlib.pylab as plt
from tensorflow import keras

# define the hyperparameters
EPOCH=500
BSIZE=64

# define the data
data = np.linspace(-np.pi, np.pi, 10000).reshape(10000, 1)
value = np.sin((np.pi) * 5 * data) / (5 * np.pi * data)

inputs = keras.Input(shape=(1,))

# Define the Model0
x = keras.layers.Dense(5, activation='relu')(inputs)
x = keras.layers.Dense(10, activation='relu')(x)
x = keras.layers.Dense(10, activation='relu')(x)
x = keras.layers.Dense(10, activation='relu')(x)
x = keras.layers.Dense(10, activation='relu')(x)
x = keras.layers.Dense(10, activation='relu')(x)
x = keras.layers.Dense(5, activation='relu')(x)
output = keras.layers.Dense(1)(x)
model = keras.Model(inputs=inputs, outputs=output)
model.compile(optimizer=tf.train.AdamOptimizer(0.0001),
              loss='mse',
              metrics=['mse'])
# show the summary of Model0
model.summary()

# Define the model1 with same parameters as model0 but it's shallow
x = keras.layers.Dense(190, activation='relu')(inputs)
model1_output = keras.layers.Dense(1)(x)
model1 = keras.Model(inputs=inputs, outputs=model1_output)
model1.compile(optimizer=tf.train.AdamOptimizer(0.0001), loss='mse', metrics=['mse'])
model1.summary()

model.fit(data, value, batch_size=BSIZE, epochs=EPOCH)
# predict the values
deep_predict = model.predict(data, batch_size=BSIZE)


model1.fit(data, value, batch_size=BSIZE, epochs=EPOCH)
shallow_predict = model1.predict(data, batch_size=BSIZE)

# plot the data
plt.plot(data, value)
plt.plot(data, deep_predict)
plt.plot(data, shallow_predict)
plt.legend(['target', 'model0', 'model1'])
plt.show()
```
